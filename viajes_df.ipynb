```python
import pandas as pd
```

```python
realizados = pd.read_csv('viajes23pd.csv')
realizados.head()
```

```python
contable = pd.read_csv('viajescontable23pd.csv')
contable.head()
```

```python
merged = pd.merge(contable,realizados, how = 'outer')
merged.head()
```

```python
monto_viaje=merged.groupby(['viajes23'])['monto'].sum()
monto_viaje
```

```python
viajes_df = pd.merge(merged,monto_viaje, how ='left', on ='viajes23')

viajes_df.rename(columns={'monto_y': 'monto_total_viaje', 'monto_x': 'monto_concepto'}, inplace=True)
viajes_df.head()
```

```python
viajes_df_clean = viajes_df[['contable23', 'viajes23', 'empleado', 'clave_concepto', 'concepto', 
       'partida', 'tipo', 'monto_concepto',
       'monto_sinavion', 'montoxdia_sinavion', 'duracion', 'inicio_fecha', 'fin_fecha', 'cancelado', 'destino',
       'dga', 'unidad','monto_total_viaje']]
viajes_df_clean=viajes_df_clean.loc[viajes_df_clean['cancelado'].isnull()]
viajes_df_clean.info()
```

```python
viajes_df_clean=viajes_df_clean.loc[viajes_df_clean['viajes23']!="viaje de otro año"]
```

```python
montos_unidad_partida = viajes_df_clean.groupby(['unidad', 'partida'])['monto_concepto'].sum().reset_index()
montos_unidad_partida.loc[montos_unidad_partida['unidad']=='Administración']
```

```python
montos_unidad_partida
```

```python
monto_total_unidad = viajes_df_clean.groupby(['unidad'])['monto_total_viaje'].sum()
duracion_total_unidad = viajes_df_clean.groupby(['unidad'])['duracion'].sum()

promedio_por_dia_unidad = monto_total_unidad/duracion_total_unidad
montos_promedio = pd.DataFrame(promedio_por_dia_unidad).rename(columns={0:'monto_concepto'})
montos_promedio['partida']='monto promedio diario'
montos_promedio = montos_promedio.reset_index()
montos_promedio = montos_promedio[['unidad','partida','monto_concepto']]
montos_promedio
```

```python
spider_df = pd.concat([montos_unidad_partida, montos_promedio])
spider_df
```
